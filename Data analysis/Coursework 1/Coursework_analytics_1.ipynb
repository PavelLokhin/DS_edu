```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from datetime import datetime, timedelta
```

```python
course_cont_df = pd.read_csv('data\course_contents.csv')
courses_df = pd.read_csv('data\courses.csv')
progresses_phases_df = pd.read_csv('data\progress_phases.csv')
progresses_df = pd.read_csv('data\progresses.csv')
students_df = pd.read_csv('data\students.csv')
```

```python
courses_df.pop('Unnamed: 0')

courses_df.rename(columns={'id':'course_id', 'title':'course_title'}, inplace=True)
progresses_df.rename(columns={'id':'progress_id'}, inplace=True)

#объединение датафрэймов
prog_merged = pd.merge(progresses_df, progresses_phases_df, on='progress_id')
course_merged = pd.merge(courses_df, course_cont_df, on='course_id')
main = pd.merge(prog_merged, course_merged)

main[main.student_id == '768c2987a744c51ce64a5993a2a94eaf']
```

```python
#общее количество курсов в датасете
courses_count = courses_df['course_title'].count()
#courses_count = len(main['course_title'].unique())
print('общее количество курсов в датасете')
print(courses_count)
print('--------------')
```

```python
#количество модулей на каждом курсе
print('количество модулей на каждом курсе')
module_per_course = course_merged.groupby('course_title')['module_title'].nunique()
print(module_per_course)
print('--------------')
```

```python
#количество уроков в каждом модуле на каждом курсе
print('количество уроков в каждом модуле на каждом курсе')
lessons_per_module = course_merged.groupby(['course_title', 'module_title'])['lesson_title'].nunique().to_frame().reset_index()
lessons_per_module.rename(columns={'lesson_title':'lesson_count'}, inplace=True)
lessons_per_module
```

```python
print('--------------')
#медианное количество уроков в модуле на каждом курсе
print('медианное количество уроков в модуле на каждом курсе')
median_les_count = lessons_per_module.groupby(['course_title'])['lesson_count'].median()
median_les_count
```

```python
print('--------------')
#количество учеников на каждом курсе
print('количество учеников на каждом курсе')
student_count = main.groupby(['course_title'])['student_id'].nunique()
print(student_count)
```

```python
current_date = datetime.now().date()
#возраст учеников
students_df['birthday'] = students_df['birthday'].fillna(value=current_date)
students_df['age'] = students_df['birthday'].apply(lambda x:
                            int((current_date - datetime.strptime(str(x), '%Y-%m-%d').date()) / timedelta(days = 365)))
#минимальный, максимальный, средний, медианный возраст студентов
#в датах есть пропуски и часть дат указано некоректно
stundet_age_min = students_df['age'][(students_df['age']>10) & (students_df['age']<90)].min()
stundet_age_max = students_df['age'][(students_df['age']>10) & (students_df['age']<90)].max()
stundet_age_mean = students_df['age'][(students_df['age']>10) & (students_df['age']<90)].mean()
stundet_age_median = students_df['age'][(students_df['age']>10) & (students_df['age']<90)].median()
print('Минимальный возраст студентов: {}, Максимальный возраст студентов: {}, Средний возраст студентов: {},'
      ' Медианный возраст студентов: {}'.format(stundet_age_min, stundet_age_max, stundet_age_mean, stundet_age_median))
```

```python
print('--------------')
#минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
print('минимальный, максимальный, средний, медианный возраст студентов на каждом курсе')
main_students = pd.merge(main, students_df, left_on='student_id', right_on='id')
main_students_grouped = main_students[(main_students['age']>10) & (main_students['age']<100)].groupby(['course_title'])\
    .agg({'age': ['min', 'max', np.average, np.median]})#в датах есть пропуски и часть дат указано некоректно
main_students_grouped
```

```python
#print(main.groupby(['course_title']).agg({'student_id':lambda x: x.nunique()}))
#количество студентов на курс
students_pcourse = main.groupby(['course_title'])['student_id'].nunique()#.agg({'student_id':lambda x: x.nunique()})
plt.title('Количество студентов на каждом курсе')
plt.tick_params(axis='x', rotation=90)
plt.bar(students_pcourse.index.values, students_pcourse.values)
plt.show()
```

```python
fig, ax = plt.subplots()

#сфера (список со сферами для последующего сопоставления и выборки списка курсов)
field = main['field'].unique()

#словарь: ключ - цвет; значение - курс 
course_color = {'r':main['course_title'][main['field']==field[0]].unique(),
               'b':main['course_title'][main['field']==field[1]].unique(),
               'y':main['course_title'][main['field']==field[2]].unique(),
               'g':main['course_title'][main['field']==field[3]].unique()}

ax.barh(students_pcourse.index.values, students_pcourse.values, align='center',
        label = field, edgecolor = 'none', alpha = 0.1)

#Заголовок
plt.title('Количество студентов на каждом курсе')
i = 0

for index, value in students_pcourse.iteritems():
    
    #выбор цвета столбцов графика в соответствии с данными из словаря course_color
    color = [col for col, course in course_color.items() if index in course]
    
    #применение цвета
    ax.patches[i].set_color(color[0])
    i += 1
    
#цвет столбцов для легенды
r_patch = mpatches.Patch(color='r', label=field[0])
b_patch = mpatches.Patch(color='b', label=field[1])
y_patch = mpatches.Patch(color='y', label=field[2])
g_patch = mpatches.Patch(color='g', label=field[3])

#медиана
plt.axvline(students_pcourse.median(), color = 'orange', linestyle='--')

#убрать рамки у графика
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
#fig.patch.set_visible(False)
#ax.axis('off')

#легенда
ax.legend(handles=[r_patch, b_patch, y_patch, g_patch])
plt.show()
```

```python
#1

#Начало обучения (в первом модуле курса 'Веб-дизайн Базовый' нет домашнего задания, 
#поэтому было решено считать начало обучения с 1-го урока 2-го модуля)
begin_study_df = main[((main['course_title']=='Веб-дизайн Базовый') & (main['module_number']==2)
                      & (main['lesson_number']==1))
                      | ((main['module_number']==1) & (main['lesson_title']=='Домашняя работа'))]

#формирование колонки с датой начала в формате Месяц Год
begin_study_df['start_date'] = pd.to_datetime(begin_study_df['start_date'])
begin_study_df['start_month'] = pd.to_datetime(begin_study_df['start_date'].apply(lambda x: datetime.strftime(x,'%Y-%m')))

#сгруппировать по дате и посчитать количество студентов по каждому периоду
begin_study_grouped = begin_study_df.groupby(['start_month','course_title'])['student_id'].count().unstack(fill_value=0).stack().to_frame().reset_index()
begin_study_grouped.head(20)
```

```python
hwork_df = main[main['lesson_title'] == 'Домашняя работа']
hwork_df_grouped = hwork_df.groupby(['course_title', 'module_number', 'module_title']).agg({'start_date' : 'count', 'finish_date' : 'count'}).reset_index()
hwork_df_grouped.head()
```

```python
#список курсов
course_list_ = hwork_df_grouped['course_title'].unique()

#Функция расчёта конверсии
def conversion(df=hwork_df_grouped, cl = course_list_):
    conv = []
    #цикл по курсам
    for item in cl:
        df_s = df[df['course_title'] == item]#выбор всех модулей указнного курса
        for index in range(len(df_s)):
            if index != 0:#у первого модуля конверсия = 0
                try:
                    conv.append(df_s['start_date'].iloc[index]/df_s['finish_date'].iloc[index-1])
                except:
                    conv.append(0)
            else:
                conv.append(0)
    return conv

hwork_df_grouped['conversion'] = conversion()
hwork_df_grouped.head()
```
